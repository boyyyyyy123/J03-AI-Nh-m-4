```python
# Import các thư viện cần thiết
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt
from sklearn_som.som import SOM # Lưu ý: Cần cài đặt sklearn_som qua pip
# Đọc dữ liệu từ file mới
df = pd.read_csv('/Users/mac/Desktop/BTL-AI1/mack.csv')
```

```python
# Define a function to calculate EMA
def calculate_ema(df, window):
    alpha = 2 / (window + 1)
    ema = [df.iloc[0]]  # Start with the first df point
    for price in df.iloc[1:]:
        ema.append(alpha * price + (1 - alpha) * ema[-1])
    return pd.Series(ema, index=df.index)

# Calculate EMA100
df['EMA100'] = calculate_ema(df['close'], 100)

# Calculate RDP values
df['RDP-5'] = (df['close'] - df['close'].shift(5)) / df['close'].shift(5) * 100
df['RDP-10'] = (df['close'] - df['close'].shift(10)) / df['close'].shift(10) * 100
df['RDP-15'] = (df['close'] - df['close'].shift(15)) / df['close'].shift(15) * 100
df['RDP-20'] = (df['close'] - df['close'].shift(20)) / df['close'].shift(20) * 100
df['RDP+5'] = (df['close'].shift(-5) - df['close']) / df['close'] * 100
```

```python
from sklearn.model_selection import GridSearchCV

# Định nghĩa các giá trị cho Grid Search
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [0.001, 0.01, 0.1, 1, 10]
}

# Khởi tạo mô hình SVM
svm_model = SVR(kernel='rbf')

# Thực hiện Grid Search với cross-validation
grid_search = GridSearchCV(svm_model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Lấy các tham số tốt nhất
best_C = grid_search.best_params_['C']
best_gamma = grid_search.best_params_['gamma']
print(f"Best C: {best_C}, Best gamma: {best_gamma}")
```

```python
# Train the SVM model with the best parameters
svm_model = SVR(kernel='rbf', C= best_C, gamma= best_gamma)
svm_model.fit(X_train, y_train)
```

```python
# Get support vectors and their corresponding coefficients
support_vectors = svm_model.support_vectors_
dual_coef = svm_model.dual_coef_[0]
```

```python
# Combine support vectors and their coefficients into a single array for clustering
support_vectors_combined = np.hstack((support_vectors, dual_coef.reshape(-1, 1)))
```

```python
from sklearn.cluster import KMeans

# Áp dụng K-Means để phân cụm các vector hỗ trợ và các hệ số
n_clusters_options = [3, 5, 7]  # Thử các số cụm khác nhau
best_mse = float('inf')
best_n_clusters = 0

for n_clusters in n_clusters_options:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    support_vector_clusters = kmeans.fit_predict(support_vectors_combined)
    
    # Lấy các trung tâm cụm
    cluster_centers = kmeans.cluster_centers_
    means = cluster_centers[:, :-1]
    sigmas = np.std(support_vectors, axis=0)
```

```python
# Hàm tính toán Gaussian membership
def gaussian_membership(x, mean, sigma):
    return np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))

# Hàm tính toán các membership
def compute_memberships(X, means, sigmas):
    memberships = np.zeros((X.shape[0], means.shape[0]))
    for i in range(means.shape[0]):
        membership = np.ones(X.shape[0])
        for j in range(X.shape[1]):
            membership *= gaussian_membership(X[:, j], means[i, j], sigmas[j])
        memberships[:, i] = membership
    return memberships

# Hàm dự đoán với Fuzzy SVM
def fuzzy_svm_predict(models, X, fuzzy_memberships):
    predictions = []
    for i in range(X.shape[0]):
        cluster_predictions = np.array([models[cluster].predict([X[i]])[0] for cluster in range(n_clusters)])
        weighted_prediction = np.dot(cluster_predictions, fuzzy_memberships[i]) / np.sum(fuzzy_memberships[i])
        predictions.append(weighted_prediction)
    return np.array(predictions)

```

```python
def loss_function(X_val, y_val, means, sigmas):
    fuzzy_memberships_val = compute_memberships(X_val, means, sigmas)
    y_pred = fuzzy_svm_predict(models, X_val, fuzzy_memberships_val)
    return mean_squared_error(y_val, y_pred)
```

```python
# Huấn luyện mô hình SVM cho mỗi cụm vector hỗ trợ
models = {}
for cluster in range(n_clusters):
    cluster_indices = np.where(support_vector_clusters == cluster)[0]
    X_cluster = support_vectors[cluster_indices]
    y_cluster = dual_coef[cluster_indices]
    
    model = SVR(kernel='rbf', C=10, gamma=0.1)
    model.fit(X_cluster, y_cluster)
    models[cluster] = model
```

```python
learning_rate = 0.001
n_iterations = 1000

for iteration in range(n_iterations):
    # Tính toán gradient của hàm mất mát với các tham số mean và sigma
    grad_means = np.zeros_like(means)
    grad_sigmas = np.zeros_like(sigmas)
    
    for i in range(means.shape[0]):
        for j in range(means.shape[1]):
            # Gradient của mean
            epsilon = 1e-5
            means[i, j] += epsilon
            loss_plus = loss_function(X_val, y_val, means, sigmas)
            means[i, j] -= 2 * epsilon
            loss_minus = loss_function(X_val, y_val, means, sigmas)
            means[i, j] += epsilon
            grad_means[i, j] = (loss_plus - loss_minus) / (2 * epsilon)
            
            # Gradient của sigma
            sigmas[j] += epsilon
            loss_plus = loss_function(X_val, y_val, means, sigmas)
            sigmas[j] -= 2 * epsilon
            loss_minus = loss_function(X_val, y_val, means, sigmas)
            sigmas[j] += epsilon
            grad_sigmas[j] = (loss_plus - loss_minus) / (2 * epsilon)
    
    # Cập nhật các tham số mean và sigma
    means -= learning_rate * grad_means
    sigmas -= learning_rate * grad_sigmas

# Tính toán các giá trị thành viên mờ cuối cùng và dự đoán kết quả
fuzzy_memberships_val = compute_memberships(X_val, means, sigmas)
y_pred = fuzzy_svm_predict(models, X_val, fuzzy_memberships_val)
mse = mean_squared_error(y_val, y_pred)
if mse < best_mse:
        best_mse = mse
        best_n_clusters = n_clusters

print(f"Best number of clusters: {best_n_clusters}, with MSE: {best_mse}")
```
